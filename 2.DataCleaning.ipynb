## Score

```python
data[data['score'].isnull()]
```

```python
data.loc[62, 'score'] = 'W/O'
data.loc[36652, 'score'] = '6-2 6-2'
data.loc[40983, 'score'] = '6-3 6-3'
data.loc[41209, 'score'] = 'W/O'
data.loc[42218, 'score'] = 'W/O'
data.loc[44403, 'score'] = 'W/O'
```

```python
# RET - retire
# W/O - walkover
# DEF - disqualification
# ABD - retire
data.loc[data['score'].str.contains('RET'), 'score']='RET' 
data.loc[data['score'].str.contains('DEF'), 'score']='DEF'
data.loc[data['score'].str.contains('ABD'), 'score']='RET'
data.loc[data['score'].str.contains('Def'), 'score']='DEF'
```

```python
# remove tie breaks scores
data['score'] = data['score'].str.replace(r'\([1-9][0-9]\)', '', regex=True)
data['score'] = data['score'].str.replace(r'\([0-9]\)', '', regex=True)
data['score'] = data['score'].str.replace(r'\[.+\]', '1-0', regex=True)
```
