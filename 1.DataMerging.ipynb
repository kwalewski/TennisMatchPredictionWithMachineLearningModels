# Data merging

```python
import pandas as pd
pd.options.display.max_columns = None
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
all_matches =  pd.DataFrame()
all_nulls = pd.DataFrame()

for i in range(1970,2024):
    df = pd.read_csv('dane/atp_matches_%s.csv'%(i))
    all_matches = pd.concat([all_matches, df], axis=0)
    
    nulls = pd.DataFrame(df.isnull().sum(), columns=['%s'%(i)])
    all_nulls = pd.concat([all_nulls, nulls], axis=1)
```

```python
#removing Davis cup matches
all_matches = all_matches[~all_matches['tourney_name'].str.contains('Davis')]
```

```python
#sorting by dates
all_matches = all_matches.sort_values(['tourney_date','match_num'])
```
